```python
import folium
# pip install folium, xyzservices, branca
```

```python
icon_lat = 60.39
icon_lon = 5.32
icon_desc = "Bergen?"
```

```python
map = folium.Map(location=[60.39, 5.32], zoom_start=4, control_scale=True)
```

```python
folium.Marker(
    location=[icon_lat, icon_lon],
    popup=icon_desc,
    icon=folium.Icon(color="green", icon="ok-sign"),
).add_to(map)
```

```python
map
```

```python
output_map = "base_map.html"
map.save(output_map)
```
